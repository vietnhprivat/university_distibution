# Videre gående uddannelser data cleaning

## Importing libraries and data

```python
# Importing libraries
import pandas as pd
import numpy as np
```

### Importing gender distibution

```python
# Read CSV local. Defining filelocation
csv_file = r"gender\accepted_gender_2022.csv"
csv_file = csv_file.replace('\\', '/')

# Creating dataframe køn from file
df_køn = pd.read_csv(csv_file, skiprows=3)


# Display dataframe from local file
display(df_køn.head(5))
display(df_køn.tail(5))

```

### Importing application requirements

```python
# Read CSV local. Defining filelocation
csv_gym = r"acceptance\accepted_acceptance_2022.csv"
csv_gym = csv_gym.replace('\\', '/')


# Creating dataframe køn from file
df_gym = pd.read_csv(csv_gym, skiprows=19)


# Display dataframe from local file
display(df_gym.head(5))
display(df_gym.tail(5))
```

## Data cleaning

### Cleaning gender distibution

```python
# Replacing studiestart... with emptyspace
df_køn['OptNavn'] = df_køn['OptNavn'].str.replace(', Studiestart: sommerstart', '')
df_køn['OptNavn'] = df_køn['OptNavn'].str.replace(', Studiestart: vinterstart', '')
df_køn['OptNavn'] = df_køn['OptNavn'].str.replace(', Studiestart: sommer- og vinterstart', '')

# Displaying dataframe
display(df_køn.head(5))
display(df_køn.tail(5))
```

### Cleaning application requirements

```python
# Changing column names to correspond with the adjacent acceptence reasoning
df_gym = df_gym.rename(columns={
    '1.0': 'STX',
    '2.0': 'HF',
    '3.0': 'HHX',
    '4.0': 'HTX',
    '5.0': 'IB',
    '6.0': 'GIF',
    '7.0': 'AFI',
    '8.0': 'EUD',
    '9.0': 'Other',
    '10.0': 'IBU',
    '11.0': 'Abroad',
    '12.0': 'EUX',
    '13.0': 'EUX_1',
    '14.0': 'Faroese',
    '15.0': 'Greenlandic'
})

# Replacing studiestart... with emptyspace
df_gym['OptNavn'] = df_gym['OptNavn'].str.replace(', Studiestart: sommerstart', '')
df_gym['OptNavn'] = df_gym['OptNavn'].str.replace(', Studiestart: vinterstart', '')
df_gym['OptNavn'] = df_gym['OptNavn'].str.replace(', Studiestart: sommer- og vinterstart', '')



# Displaying dataframe
display(df_gym.head(5))
display(df_gym.tail(5))
```

## File Export

```python
# Exporting dataframe as csv file 
df_køn.to_csv('df_køn.csv', index=False)
```

```python
# Exporting gym to csv
df_gym.to_csv('gym.csv', index=False)
```
